```python
# load dataset
dataset = pd.read_excel("/content/drive/MyDrive/Doutorado/HELV/Metricas_VFC_Covid_Modificado_bin.xlsx")
#dataset = pd.read_excel("/content/drive/MyDrive/Doutorado/HELV/Metricas_VFC_Covid_Modificado.xlsx")
dataset.head()
```

```python
dataset = dataset.dropna()
```

```python
dataset.head()
```

```python
#y = np.array(dataset["gravidade"], dtype=int)
y = np.array(dataset["gravidade"])
```

```python
from sklearn.preprocessing import LabelBinarizer
y = LabelBinarizer().fit_transform(y)
```

```python
X = np.array(dataset)
X = X[:,:44]
X = np.array(X, dtype=float)
```

```python
from sklearn import preprocessing
X = preprocessing.normalize(X, norm='l2')
```

```python
#prepare models
modelos = [
    SVC(gamma='auto'),
#   SVC(kernel="linear", C=0.025),
#   SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    LinearSVC(),
    SGDClassifier(max_iter=100, tol=1e-3), 
    KNeighborsClassifier(),
    LogisticRegression(solver='lbfgs'), 
    LogisticRegressionCV(cv=3),
    BaggingClassifier(), 
    ExtraTreesClassifier(n_estimators=300),
    RandomForestClassifier(max_depth=5, n_estimators=300, max_features=1),
    GaussianNB(), 
    DecisionTreeClassifier(max_depth=5),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis(),
    OneVsRestClassifier(LinearSVC(random_state=0)) #multiclass
]
```

```python
# prepare configuration for cross validation test harness
seed = 10
splits = 10
qtd_modelos = 0
algoritmos = []
acuracia = []
roc_auc = []
revogacao = []
precisao = []
f1 = []
resultados = pd.DataFrame(columns=['algoritmo','acurácia', 'roc auc', 'revogação', 'precisão', 'f1'])
metricas = ['accuracy', 'roc_auc', 'recall', 'precision', 'f1']
for modelo in modelos:
	print(f"Processando {modelo.__class__.__name__}")
	qtd_modelos += 1
	kfold = model_selection.KFold(n_splits=splits, random_state=seed, shuffle=True)
	algoritmos.append(modelo.__class__.__name__)
	qual_metrica = 0
	for metrica in metricas:
		cv_results = model_selection.cross_val_score(modelo, X, y, cv=kfold, scoring=metrica)
		if qual_metrica == 0:
			acuracia.append(cv_results.mean())
		if qual_metrica == 1:
			roc_auc.append(cv_results.mean())
		if qual_metrica == 2:
			revogacao.append(cv_results.mean())
		if qual_metrica == 3:
			precisao.append(cv_results.mean())
		if qual_metrica == 4:
			f1.append(cv_results.mean())
		qual_metrica += 1
	
```

```python
#montando dataframe
resultados['algoritmo'] = algoritmos
#resultados['roc_auc'] = roc_auc
resultados['acurácia'] = acuracia
resultados['revogação'] = revogacao
resultados['precisão'] = precisao
resultados['f1'] = f1
```
